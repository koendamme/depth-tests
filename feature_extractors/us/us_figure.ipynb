```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from extract_us_wave import get_wave_updated
```

```python
with open("surrogates.pickle", "rb") as file:
    us = np.array(pickle.load(file)["us"]).T
```

```python

fig, ax = plt.subplots()
ax.plot(us[300:800, 200])
# ax.set_xlabel("Depth")
# ax.set_ylabel("Signal strength")
# ax.spines['left'].set_position('zero')
# # turn off the right spine/ticks
# ax.spines['right'].set_color('none')
# ax.yaxis.tick_left()
#
# # set the y-spine
# ax.spines['bottom'].set_position('zero')
#
# # turn off the top spine/ticks
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
# ax.xaxis.tick_bottom()

ax.set_xticks([])
ax.set_yticks([0])
plt.tight_layout()
plt.savefig(f"raw.png")
```
